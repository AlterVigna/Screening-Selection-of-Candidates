```python
## Funtion to pass from single specific skill to macrogroup of skills.
def aggregateSkills(df):
    df['Backend'] = np.where((df["cpp"]==1) | (df[".net"]==1) |  (df["python"]==1) | 
                  (df["java"]==1) | (df["ruby"]==1) | (df["react"]==1) | (df["php"]==1), 1, 0)
    df["Frontend"]=np.where((df["html"]==1) | (df["css"]==1) | (df["javascript"]==1) , 1, 0)
    df["MiddleWare"]=np.where((df["angular"]==1) | (df["rabbitmq"]==1) | (df["kubernetes"]==1), 1, 0)
    df["OS"]=np.where((df["angular"]==1) | (df["rabbitmq"]==1) | (df["kubernetes"]==1) , 1, 0)
    df["Database"]=np.where((df["sql"]==1) | (df["nosql"]==1) | (df["mongodb"]==1) | (df["oracle"]==1) | (df["neo4j"]==1), 1, 0)
    df["AI"]=np.where((df["ai"]==1) | (df["tensorflow"]==1) | (df["nlp"]==1), 1, 0)
    df["Networks"]=np.where((df["networking"]==1) | (df["security"]==1) | (df["cloud"]==1), 1, 0)
    df["General"]=np.where((df["test"]==1) | (df["team"]==1), 1, 0)
    df['Total']=df['Backend']
    return df
```
